```python
import pandas as pd
from scipy import stats
```

```python
populationDf = pd.read_excel('input/DDW_PCA0000_2011_Indiastatedist.xlsx')
populationDf
```

```python
populationDf = populationDf[['State','Level','Name','TRU','TOT_M','TOT_F']]
populationDf = populationDf[populationDf['TRU']=='Total'].drop(['TRU'], axis=1)
populationDf['Name'] = populationDf['Name'].str.upper()
populationDf = populationDf[(populationDf['Level']=='India') | (populationDf['Level']=='STATE')].drop(['Level'], axis=1).reset_index(drop=True)
populationDf
```

```python
languageDf = pd.read_excel('input/DDW-C18-0000.xlsx', names=['state-code','District','State Name','TRU','Age_Group','Persons_L2','Males_L2','Females_L2','Persons_L3','Males_L3','Females_L3'])
languageDf = languageDf[languageDf['TRU']=='Total'].drop(['TRU'], axis=1)
languageDf = languageDf[languageDf['Age_Group']=='Total'].drop(['State Name','District','Age_Group','Persons_L2','Persons_L3'], axis=1).reset_index(drop=True)
languageDf = languageDf.astype({'state-code':int})
languageDf
```

```python
languageDf['Only1LangM'] = populationDf['TOT_M'] - languageDf['Males_L2']
languageDf['Only1LangF'] = populationDf['TOT_F'] - languageDf['Females_L2']
languageDf['Exact2LangM'] = languageDf['Males_L2'] - languageDf['Males_L3']
languageDf['Exact2LangF'] = languageDf['Females_L2'] - languageDf['Females_L3']
languageDf['Lang3orMoreM'] = languageDf['Males_L3']
languageDf['Lang3orMoreF'] = languageDf['Females_L3']
languageDf['ratioL1'] = languageDf['Only1LangM']/languageDf['Only1LangF']
languageDf['ratioL2'] = languageDf['Exact2LangM']/languageDf['Exact2LangF']
languageDf['ratioL3'] = languageDf['Lang3orMoreM']/languageDf['Lang3orMoreF']
languageDf['ratioPopulation'] = populationDf['TOT_M']/populationDf['TOT_F']
languageDf
```

```python
languageDf1 = languageDf[['state-code']]
languageDf1['male-percentage'] = (languageDf['Only1LangM']/populationDf['TOT_M'])*100
languageDf1['female-percentage'] = (languageDf['Only1LangF']/populationDf['TOT_F'])*100
languageDf1['p-value'] = languageDf.apply(lambda r: stats.ttest_1samp([r.ratioL1,r.ratioL2,r.ratioL3], popmean=r.ratioPopulation)[1], axis=1)
languageDf1
```

```python
languageDf2 = languageDf[['state-code']]
languageDf2['male-percentage'] = (languageDf['Exact2LangM']/populationDf['TOT_M'])*100
languageDf2['female-percentage'] = (languageDf['Exact2LangF']/populationDf['TOT_F'])*100
languageDf2['p-value'] = languageDf.apply(lambda r: stats.ttest_1samp([r.ratioL1,r.ratioL2,r.ratioL3], popmean=r.ratioPopulation)[1], axis=1)
languageDf2
```

```python
languageDf3 = languageDf[['state-code']]
languageDf3['male-percentage'] = (languageDf['Lang3orMoreM']/populationDf['TOT_M'])*100
languageDf3['female-percentage'] = (languageDf['Lang3orMoreF']/populationDf['TOT_F'])*100
languageDf3['p-value'] = languageDf.apply(lambda r: stats.ttest_1samp([r.ratioL1,r.ratioL2,r.ratioL3], popmean=r.ratioPopulation)[1], axis=1)
languageDf3
```

```python
languageDf1.to_csv('gender-india-a.csv', index=False)
languageDf2.to_csv('gender-india-b.csv', index=False)
languageDf3.to_csv('gender-india-c.csv', index=False)
```
