```python
import pandas as pd
```

```python
populationDf = pd.read_excel('input/DDW-0000C-14.xls', names=['tableName','State Code','District','state/ut','agegroup','Persons','Males','Females','Persons_L2','Males_L2','Females_L2','Persons_L3','Males_L3','Females_L3'])
populationDf = populationDf[['State Code','state/ut','agegroup','Males','Females']]
populationDf = populationDf.iloc[6:, : ]
populationDf = populationDf.astype({'State Code':int})
populationDf = populationDf[populationDf.agegroup != '0-4']
populationDf.head(20)
```

```python
pDf = pd.DataFrame()
for stCode in range(36):
    stDf = populationDf[populationDf['State Code'] == stCode]
    stDf = stDf.reset_index(drop=True)
    stDf.loc[6,'Males'] = stDf[(stDf.agegroup == '30-34') | (stDf.agegroup == '35-39') | (stDf.agegroup == '40-44') | (stDf.agegroup == '45-49')]['Males'].sum()
    stDf.loc[6,'Females'] = stDf[(stDf.agegroup == '30-34') | (stDf.agegroup == '35-39') | (stDf.agegroup == '40-44') | (stDf.agegroup == '45-49')]['Females'].sum()
    stDf.loc[6,'agegroup'] = '30-49'
    stDf.loc[10,'Males'] = stDf[(stDf.agegroup == '50-54') | (stDf.agegroup == '55-59') | (stDf.agegroup == '60-64') | (stDf.agegroup == '65-69')]['Males'].sum()
    stDf.loc[10,'Females'] = stDf[(stDf.agegroup == '50-54') | (stDf.agegroup == '55-59') | (stDf.agegroup == '60-64') | (stDf.agegroup == '65-69')]['Females'].sum()
    stDf.loc[10,'agegroup'] = '50-69'
    stDf.loc[14,'Males'] = stDf[(stDf.agegroup == '70-74') | (stDf.agegroup == '75-79') | (stDf.agegroup == '80+')]['Males'].sum()
    stDf.loc[14,'Females'] = stDf[(stDf.agegroup == '70-74') | (stDf.agegroup == '75-79') | (stDf.agegroup == '80+')]['Females'].sum()
    stDf.loc[14,'agegroup'] = '70+'
    stDf = stDf[(stDf.agegroup != 'All ages') & (stDf.agegroup != '35-39') & (stDf.agegroup != '40-44') & (stDf.agegroup != '45-49') & (stDf.agegroup != '55-59') & (stDf.agegroup != '60-64') & (stDf.agegroup != '65-69') & (stDf.agegroup != '75-79') & (stDf.agegroup != '80+')]
    pDf = pd.concat([pDf,stDf])

pDf = pDf.reset_index(drop=True)
pDf
```

```python
languageDf = pd.read_excel('input/DDW-C18-0000.xlsx', names=['State Code','District','state/ut','TRU','age-group','Persons_L2','Males_L2','Females_L2','Persons_L3','Males_L3','Females_L3'])
languageDf = languageDf[languageDf['TRU']=='Total'].drop(['TRU'], axis=1)
languageDf = languageDf[languageDf['age-group']!='Total'].drop(['District','Persons_L2','Persons_L3'], axis=1).reset_index(drop=True)
languageDf = languageDf.astype({'State Code':int})
languageDf
```

```python
languageDf['only1langM'] = pDf['Males'] - languageDf['Males_L2']
languageDf['only1langF'] = pDf['Females'] - languageDf['Females_L2']

languageDf['exact2langM'] = languageDf['Males_L2'] - languageDf['Males_L3']
languageDf['exact2langF'] = languageDf['Females_L2'] - languageDf['Females_L3']

languageDf['lang3ormoreM'] = languageDf['Males_L3']
languageDf['lang3ormoreF'] = languageDf['Females_L3']

languageDf['ratioL1M'] = languageDf['only1langM']/pDf['Males']
languageDf['ratioL1F'] = languageDf['only1langF']/pDf['Females']
languageDf['ratioL2M'] = languageDf['exact2langM']/pDf['Males']
languageDf['ratioL2F'] = languageDf['exact2langF']/pDf['Females']
languageDf['ratioL3M'] = languageDf['lang3ormoreM']/pDf['Males']
languageDf['ratioL3F'] = languageDf['lang3ormoreF']/pDf['Females']

languageDf
```

```python
ageIndiaL3Df = pd.DataFrame(columns=['state/ut','age-group-males','ratio-males','age-group-females','ratio-females'])
for stCode in range(36):
    stDf = languageDf[languageDf['State Code'] == stCode].reset_index(drop=True)
    stMDf = stDf.sort_values(by=['ratioL3M'], ascending=False).reset_index(drop=True)
    stMDf = stMDf.head(1)
    stFDf = stDf.sort_values(by=['ratioL3F'], ascending=False).reset_index(drop=True)
    stFDf = stFDf.head(1)
    ageIndiaL3Df.loc[len(ageIndiaL3Df.index)] = [stCode,stMDf.iloc[0]['age-group'],stMDf.iloc[0]['ratioL3M'],stFDf.iloc[0]['age-group'],stFDf.iloc[0]['ratioL3F']]

ageIndiaL3Df
```

```python
ageIndiaL2Df = pd.DataFrame(columns=['state/ut','age-group-males','ratio-males','age-group-females','ratio-females'])
for stCode in range(36):
    stDf = languageDf[languageDf['State Code'] == stCode].reset_index(drop=True)
    stMDf = stDf.sort_values(by=['ratioL2M'], ascending=False).reset_index(drop=True)
    stMDf = stMDf.head(1)
    stFDf = stDf.sort_values(by=['ratioL2F'], ascending=False).reset_index(drop=True)
    stFDf = stFDf.head(1)
    ageIndiaL2Df.loc[len(ageIndiaL2Df.index)] = [stCode,stMDf.iloc[0]['age-group'],stMDf.iloc[0]['ratioL2M'],stFDf.iloc[0]['age-group'],stFDf.iloc[0]['ratioL2F']]

ageIndiaL2Df
```

```python
ageIndiaL1Df = pd.DataFrame(columns=['state/ut','age-group-males','ratio-males','age-group-females','ratio-females'])
for stCode in range(36):
    stDf = languageDf[languageDf['State Code'] == stCode].reset_index(drop=True)
    stMDf = stDf.sort_values(by=['ratioL1M'], ascending=False).reset_index(drop=True)
    stMDf = stMDf.head(1)
    stFDf = stDf.sort_values(by=['ratioL1F'], ascending=False).reset_index(drop=True)
    stFDf = stFDf.head(1)
    ageIndiaL1Df.loc[len(ageIndiaL1Df.index)] = [stCode,stMDf.iloc[0]['age-group'],stMDf.iloc[0]['ratioL1M'],stFDf.iloc[0]['age-group'],stFDf.iloc[0]['ratioL1F']]

ageIndiaL1Df
```

```python
ageIndiaL3Df.to_csv('age-gender-a.csv',index=False)
ageIndiaL2Df.to_csv('age-gender-b.csv',index=False)
ageIndiaL1Df.to_csv('age-gender-c.csv',index=False)
```
