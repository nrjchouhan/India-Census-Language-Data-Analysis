```python
import pandas as pd
```

```python
populationDf = pd.read_excel('input/DDW_PCA0000_2011_Indiastatedist.xlsx')
populationDf = populationDf[['State','Level','Name','TRU','TOT_P']]
populationDf = populationDf[populationDf['TRU']=='Total'].drop(['TRU'], axis=1)
populationDf['Name'] = populationDf['Name'].str.upper()
populationDf = populationDf[(populationDf['Level']=='India') | (populationDf['Level']=='STATE')].drop(['Level'], axis=1).reset_index(drop=True)
populationDf
```

```python
languageDf = pd.read_excel('input/DDW-C18-0000.xlsx', names=['state-code','District','State Name','TRU','Age_Group','Persons_L2','Males_L2','Females_L2','Persons_L3','Males_L3','Females_L3'])
languageDf = languageDf[languageDf['TRU']=='Total'].drop(['TRU'], axis=1)
languageDf = languageDf[languageDf['Age_Group']=='Total'].drop(['District','Age_Group','Males_L2','Females_L2','Males_L3','Females_L3'], axis=1).reset_index(drop=True)
languageDf = languageDf.astype({'state-code':int})
languageDf
```

```python
languageDf['Only1Lang'] = populationDf['TOT_P'] - languageDf['Persons_L2']
languageDf['Exact2Lang'] = languageDf['Persons_L2'] - languageDf['Persons_L3']
languageDf['Lang3orMore'] = populationDf['TOT_P'] - languageDf['Only1Lang'] - languageDf['Exact2Lang']
languageDf
```

```python
languageDf['percent-one'] = (languageDf['Only1Lang']/populationDf['TOT_P'])*100
languageDf['percent-two'] = (languageDf['Exact2Lang']/populationDf['TOT_P'])*100
languageDf['percent-three'] = (languageDf['Lang3orMore']/populationDf['TOT_P'])*100
languageDf = languageDf.drop(['State Name','Persons_L2','Persons_L3','Only1Lang','Exact2Lang','Lang3orMore'], axis=1)
languageDf
```

```python
languageDf.to_csv('percent-india.csv', index=False)
```
