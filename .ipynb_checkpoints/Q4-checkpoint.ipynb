```python
import pandas as pd
```

```python
percentIndiaDf = pd.read_csv('percent-india.csv')
percentIndiaDf
```

```python
lang3To2Df = percentIndiaDf[['state-code']]
lang3To2Df['3-to-2-ratio'] = percentIndiaDf['percent-three']/percentIndiaDf['percent-two']
lang3To2Df = lang3To2Df.sort_values(by=['3-to-2-ratio'], ascending=False).reset_index(drop=True)
best3St3To2Df = lang3To2Df.head(3)
worst3St3To2Df = lang3To2Df.tail(3).sort_values(by=['3-to-2-ratio']).reset_index(drop=True)
lang3To2BestWorstDf = pd.concat([best3St3To2Df,worst3St3To2Df], axis=0)
lang3To2BestWorstDf
```

```python
lang3To2BestWorstDf = lang3To2BestWorstDf[['state-code']]
lang3To2BestWorstDf
```

```python
lang2To1Df = percentIndiaDf[['state-code']]
lang2To1Df['2-to-1-ratio'] = percentIndiaDf['percent-two']/percentIndiaDf['percent-one']
lang2To1Df = lang2To1Df.sort_values(by=['2-to-1-ratio'], ascending=False).reset_index(drop=True)
best3St2To1Df = lang2To1Df.head(3)
worst3St2To1Df = lang2To1Df.tail(3).sort_values(by=['2-to-1-ratio']).reset_index(drop=True)
lang2To1BestWorstDf = pd.concat([best3St2To1Df,worst3St2To1Df], axis=0)
lang2To1BestWorstDf
```

```python
lang2To1BestWorstDf = lang2To1BestWorstDf[['state-code']]
lang2To1BestWorstDf
```

```python
lang3To2BestWorstDf.to_csv('3-to-2-ratio.csv', index=False)
lang2To1BestWorstDf.to_csv('2-to-1-ratio.csv', index=False)
```
