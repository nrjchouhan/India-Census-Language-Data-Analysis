```python
import pandas as pd
from scipy import stats
```

```python
populationDf = pd.read_excel('input/DDW_PCA0000_2011_Indiastatedist.xlsx')
populationDf
```

```python
populationDf = populationDf[['State','Level','Name','TRU','TOT_P']]
populationUrbanDf = populationDf[populationDf['TRU']=='Urban'].drop(['TRU'], axis=1)
populationUrbanDf['Name'] = populationUrbanDf['Name'].str.upper()
populationUrbanDf = populationUrbanDf[(populationUrbanDf['Level']=='India') | (populationUrbanDf['Level']=='STATE')].drop(['Level'], axis=1).reset_index(drop=True)
populationUrbanDf
```

```python
populationRuralDf = populationDf[populationDf['TRU']=='Rural'].drop(['TRU'], axis=1)
populationRuralDf['Name'] = populationRuralDf['Name'].str.upper()
populationRuralDf = populationRuralDf[(populationRuralDf['Level']=='India') | (populationRuralDf['Level']=='STATE')].drop(['Level'], axis=1).reset_index(drop=True)
populationRuralDf
```

```python
languageDf = pd.read_excel('input/DDW-C18-0000.xlsx', names=['state-code','District','State Name','TRU','Age_Group','Persons_L2','Males_L2','Females_L2','Persons_L3','Males_L3','Females_L3'])
languageUrbanDf = languageDf[languageDf['TRU']=='Urban'].drop(['TRU'], axis=1)
languageUrbanDf = languageUrbanDf[languageUrbanDf['Age_Group']=='Total'].drop(['State Name','District','Age_Group','Males_L2','Females_L2','Males_L3','Females_L3'], axis=1).reset_index(drop=True)
languageUrbanDf = languageUrbanDf.astype({'state-code':int})
languageUrbanDf
```

```python
languageRuralDf = languageDf[languageDf['TRU']=='Rural'].drop(['TRU'], axis=1)
languageRuralDf = languageRuralDf[languageRuralDf['Age_Group']=='Total'].drop(['State Name','District','Age_Group','Males_L2','Females_L2','Males_L3','Females_L3'], axis=1).reset_index(drop=True)
languageRuralDf = languageRuralDf.astype({'state-code':int})
languageRuralDf
```

```python
resultDf = languageUrbanDf[['state-code']]
resultDf['Only1LangU'] = populationUrbanDf['TOT_P'] - languageUrbanDf['Persons_L2']
resultDf['Only1LangR'] = populationRuralDf['TOT_P'] - languageRuralDf['Persons_L2']
resultDf['Exact2LangU'] = languageUrbanDf['Persons_L2'] - languageUrbanDf['Persons_L3']
resultDf['Exact2LangR'] = languageRuralDf['Persons_L2'] - languageRuralDf['Persons_L3']
resultDf['Lang3orMoreU'] = languageUrbanDf['Persons_L3']
resultDf['Lang3orMoreR'] = languageRuralDf['Persons_L3']
resultDf['ratioL1'] = resultDf['Only1LangU']/resultDf['Only1LangR']
resultDf['ratioL2'] = resultDf['Exact2LangU']/resultDf['Exact2LangR']
resultDf['ratioL3'] = resultDf['Lang3orMoreU']/resultDf['Lang3orMoreR']
resultDf['ratioPopulation'] = populationUrbanDf['TOT_P']/populationRuralDf['TOT_P']
resultDf
```

```python
resultDf1 = resultDf[['state-code']]
resultDf1['urban-percentage'] = (resultDf['Only1LangU']/populationUrbanDf['TOT_P'])*100
resultDf1['rural-percentage'] = (resultDf['Only1LangR']/populationRuralDf['TOT_P'])*100
resultDf1['p-value'] = resultDf.apply(lambda r: stats.ttest_1samp([r.ratioL1,r.ratioL2,r.ratioL3], popmean=r.ratioPopulation)[1], axis=1)
resultDf1
```

```python
resultDf2 = resultDf[['state-code']]
resultDf2['urban-percentage'] = (resultDf['Exact2LangU']/populationUrbanDf['TOT_P'])*100
resultDf2['rural-percentage'] = (resultDf['Exact2LangR']/populationRuralDf['TOT_P'])*100
resultDf2['p-value'] = resultDf.apply(lambda r: stats.ttest_1samp([r.ratioL1,r.ratioL2,r.ratioL3], popmean=r.ratioPopulation)[1], axis=1)
resultDf2
```

```python
resultDf3 = resultDf[['state-code']]
resultDf3['urban-percentage'] = (resultDf['Lang3orMoreU']/populationUrbanDf['TOT_P'])*100
resultDf3['rural-percentage'] = (resultDf['Lang3orMoreR']/populationRuralDf['TOT_P'])*100
resultDf3['p-value'] = resultDf.apply(lambda r: stats.ttest_1samp([r.ratioL1,r.ratioL2,r.ratioL3], popmean=r.ratioPopulation)[1], axis=1)
resultDf3
```

```python
resultDf1.to_csv('geography-india-a.csv', index=False)
resultDf2.to_csv('geography-india-b.csv', index=False)
resultDf3.to_csv('geography-india-c.csv', index=False)
```
