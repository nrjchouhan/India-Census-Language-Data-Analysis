```python
# Importing libraries
import pandas as pd
```

```python
# Reading population data with education level
populationDf = pd.read_excel('input/DDW-0000C-08.xlsx', names=['table','State Code','distCode','Name','TRU','agegroup','totalP','totalM','totalF','illitP','illitM','illitF','litP','litM','litF','litWoEdP','litWoEdM','litWoEdF','blwPriP','blwPriM','blwPriF','priP','priM','priF','midP','midM','midF','matrcP','matrcM','matrcF','highSecP','highSecM','highSecF','dipNTP','dipNTM','dipNTF','dipTP','dipTM','dipTF','gradP','gradM','gradF','unclassP','unclassM','unclassF','blank'])
populationDf.head(20)
```

```python
# Preprocessing populating data
populationDf = populationDf.iloc[6:, : ]
populationDf = populationDf.astype({'State Code':int})
populationDf = populationDf[populationDf['TRU']=='Total']
populationDf = populationDf[populationDf['agegroup']=='All ages']
populationDf = populationDf[['State Code','Name','illitM','illitF','blwPriM','blwPriF','priM','priF','midM','midF','matrcM','matrcF','highSecM','highSecF','dipNTM','dipNTF','dipTM','dipTF','gradM','gradF']]
populationDf
```

```python
pDf = pd.DataFrame(columns=['state/ut','literacy-group','Males','Females'])
# Grouping literacy-group so that they should be in same order as in language speakers data
for stCode in range(36):
    stDf = populationDf[populationDf['State Code']==stCode]
    pDf.loc[len(pDf.index)] = [stCode,'Illiterate',stDf.iloc[0]['illitM'],stDf.iloc[0]['illitF']]
    pDf.loc[len(pDf.index)] = [stCode,'Literate but below primary',stDf.iloc[0]['blwPriM'],stDf.iloc[0]['blwPriF']]
    pDf.loc[len(pDf.index)] = [stCode,'Primary but below middle',stDf.iloc[0]['priM'],stDf.iloc[0]['priF']]
    pDf.loc[len(pDf.index)] = [stCode,'Middle but below matric/secondary',stDf.iloc[0]['midM'],stDf.iloc[0]['midF']]
    pDf.loc[len(pDf.index)] = [stCode,'Matric/Secondary but below graduate',stDf.iloc[0]['matrcM']+stDf.iloc[0]['highSecM']+stDf.iloc[0]['dipNTM']+stDf.iloc[0]['dipTM'],stDf.iloc[0]['matrcF']+stDf.iloc[0]['highSecF']+stDf.iloc[0]['dipNTF']+stDf.iloc[0]['dipTF']]
    pDf.loc[len(pDf.index)] = [stCode,'Graduate and above',stDf.iloc[0]['gradM'],stDf.iloc[0]['gradF']]

pDf = pDf.reset_index(drop=True)
pDf
```

```python
# Reading language speakers data with literacy-group
languageDf = pd.read_excel('input/DDW-C19-0000.xlsx', names=['State Code','District','state/ut','TRU','literacy-group','Persons_L2','Males_L2','Females_L2','Persons_L3','Males_L3','Females_L3'])
languageDf = languageDf[languageDf['TRU']=='Total'].drop(['TRU'], axis=1)
languageDf = languageDf[(languageDf['literacy-group']!='Total') & (languageDf['literacy-group']!='Literate')].drop(['District','Persons_L2','Persons_L3'], axis=1).reset_index(drop=True)
languageDf = languageDf.astype({'State Code':int})
languageDf
```

```python
# Calculating num of males and felmales speak only one language, exactly two languages and three languages and more
languageDf['only1langM'] = pDf['Males'] - languageDf['Males_L2']
languageDf['only1langF'] = pDf['Females'] - languageDf['Females_L2']

languageDf['exact2langM'] = languageDf['Males_L2'] - languageDf['Males_L3']
languageDf['exact2langF'] = languageDf['Females_L2'] - languageDf['Females_L3']

languageDf['lang3ormoreM'] = languageDf['Males_L3']
languageDf['lang3ormoreF'] = languageDf['Females_L3']

# Calculating ratio of males and felmales to population speak only one language, exactly two languages and three languages and more
languageDf['ratioL1M'] = languageDf['only1langM']/pDf['Males']
languageDf['ratioL1F'] = languageDf['only1langF']/pDf['Females']
languageDf['ratioL2M'] = languageDf['exact2langM']/pDf['Males']
languageDf['ratioL2F'] = languageDf['exact2langF']/pDf['Females']
languageDf['ratioL3M'] = languageDf['lang3ormoreM']/pDf['Males']
languageDf['ratioL3F'] = languageDf['lang3ormoreF']/pDf['Females']

languageDf
```

```python
literacyIndiaL3Df = pd.DataFrame(columns=['state/ut','literacy-group-males','ratio-males','literacy-group-females','ratio-females'])
# Finding literacy-group from india and each state with highest percentage of people speaking three languages or more for males and females
for stCode in range(36):
    stDf = languageDf[languageDf['State Code'] == stCode].reset_index(drop=True)
    stMDf = stDf.sort_values(by=['ratioL3M'], ascending=False).reset_index(drop=True)
    stMDf = stMDf.head(1)
    stFDf = stDf.sort_values(by=['ratioL3F'], ascending=False).reset_index(drop=True)
    stFDf = stFDf.head(1)
    literacyIndiaL3Df.loc[len(literacyIndiaL3Df.index)] = [stCode,stMDf.iloc[0]['literacy-group'],stMDf.iloc[0]['ratioL3M'],stFDf.iloc[0]['literacy-group'],stFDf.iloc[0]['ratioL3F']]

literacyIndiaL3Df
```

```python
literacyIndiaL2Df = pd.DataFrame(columns=['state/ut','literacy-group-males','ratio-males','literacy-group-females','ratio-females'])
# Finding literacy-group from india and each state with highest percentage of people speaking exactly two languages for males and females
for stCode in range(36):
    stDf = languageDf[languageDf['State Code'] == stCode].reset_index(drop=True)
    stMDf = stDf.sort_values(by=['ratioL2M'], ascending=False).reset_index(drop=True)
    stMDf = stMDf.head(1)
    stFDf = stDf.sort_values(by=['ratioL2F'], ascending=False).reset_index(drop=True)
    stFDf = stFDf.head(1)
    literacyIndiaL2Df.loc[len(literacyIndiaL2Df.index)] = [stCode,stMDf.iloc[0]['literacy-group'],stMDf.iloc[0]['ratioL2M'],stFDf.iloc[0]['literacy-group'],stFDf.iloc[0]['ratioL2F']]

literacyIndiaL2Df
```

```python
literacyIndiaL1Df = pd.DataFrame(columns=['state/ut','literacy-group-males','ratio-males','literacy-group-females','ratio-females'])
# Finding literacy-group from india and each state with highest percentage of people speaking only one language for males and females
for stCode in range(36):
    stDf = languageDf[languageDf['State Code'] == stCode].reset_index(drop=True)
    stMDf = stDf.sort_values(by=['ratioL1M'], ascending=False).reset_index(drop=True)
    stMDf = stMDf.head(1)
    stFDf = stDf.sort_values(by=['ratioL1F'], ascending=False).reset_index(drop=True)
    stFDf = stFDf.head(1)
    literacyIndiaL1Df.loc[len(literacyIndiaL1Df.index)] = [stCode,stMDf.iloc[0]['literacy-group'],stMDf.iloc[0]['ratioL1M'],stFDf.iloc[0]['literacy-group'],stFDf.iloc[0]['ratioL1F']]

literacyIndiaL1Df
```

```python
# Generating output csv file from datframe
literacyIndiaL3Df.to_csv('literacy-gender-a.csv',index=False)
literacyIndiaL2Df.to_csv('literacy-gender-b.csv',index=False)
literacyIndiaL1Df.to_csv('literacy-gender-c.csv',index=False)
```
