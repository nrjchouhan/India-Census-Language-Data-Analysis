```python
# Importing libraries
import pandas as pd
```

```python
# Reading population data with education level
populationDf = pd.read_excel('input/DDW-0000C-08.xlsx', names=['table','State Code','distCode','Name','TRU','agegroup','totalP','totalM','totalF','illitP','illitM','illitF','litP','litM','litF','litWoEdP','litWoEdM','litWoEdF','blwPriP','blwPriM','blwPriF','priP','priM','priF','midP','midM','midF','matrcP','matrcM','matrcF','highSecP','highSecM','highSecF','dipNTP','dipNTM','dipNTF','dipTP','dipTM','dipTF','gradP','gradM','gradF','unclassP','unclassM','unclassF','blank'])
populationDf.head(20)
```

```python
# Preprocessing populating data
populationDf = populationDf.iloc[6:, : ]
populationDf = populationDf.astype({'State Code':int})
populationDf = populationDf[populationDf['TRU']=='Total']
populationDf = populationDf[populationDf['agegroup']=='All ages']
populationDf = populationDf[['State Code','Name','illitP','blwPriP','priP','midP','matrcP','highSecP','dipNTP','dipTP','gradP']]
populationDf
```

```python
pDf = pd.DataFrame(columns=['State Code','literacy-group','population'])
# Grouping literacy-group so that they should be in same order as in language speakers data
for stCode in range(36):
    stDf = populationDf[populationDf['State Code']==stCode]
    pDf.loc[len(pDf.index)] = [stCode,'Illiterate',stDf.iloc[0]['illitP']]
    pDf.loc[len(pDf.index)] = [stCode,'Literate but below primary',stDf.iloc[0]['blwPriP']]
    pDf.loc[len(pDf.index)] = [stCode,'Primary but below middle',stDf.iloc[0]['priP']]
    pDf.loc[len(pDf.index)] = [stCode,'Middle but below matric/secondary',stDf.iloc[0]['midP']]
    pDf.loc[len(pDf.index)] = [stCode,'Matric/Secondary but below graduate',stDf.iloc[0]['matrcP']+stDf.iloc[0]['highSecP']+stDf.iloc[0]['dipNTP']+stDf.iloc[0]['dipTP']]
    pDf.loc[len(pDf.index)] = [stCode,'Graduate and above',stDf.iloc[0]['gradP']]

pDf
```

```python
# Reading language speakers data with literacy-group
languageDf = pd.read_excel('input/DDW-C19-0000.xlsx', names=['State Code','District','state/ut','TRU','literacy-group','Persons_L2','Males_L2','Females_L2','Persons_L3','Males_L3','Females_L3'])
languageDf = languageDf[languageDf['TRU']=='Total'].drop(['TRU'], axis=1)
languageDf = languageDf[(languageDf['literacy-group']!='Total') & (languageDf['literacy-group']!='Literate')].drop(['District','Persons_L2','Males_L2','Females_L2','Males_L3','Females_L3'], axis=1).reset_index(drop=True)
languageDf = languageDf.astype({'State Code':int})
languageDf
```

```python
literacyIndiaDf = pd.DataFrame()
# Finding literacy-group from india and each state with highest percentage of people speaking three languages or more
for stCode in range(36):
    stDf = languageDf[languageDf['State Code'] == stCode].reset_index(drop=True)
    pStDf = pDf[pDf['State Code'] == stCode].reset_index(drop=True)
    stDf['percentage'] = (stDf['Persons_L3']/pStDf['population'])*100
    stDf = stDf.sort_values(by=['percentage'], ascending=False).reset_index(drop=True)
    literacyIndiaDf = pd.concat([literacyIndiaDf,stDf.head(1)])
literacyIndiaDf = literacyIndiaDf[['State Code','literacy-group','percentage']]
literacyIndiaDf = literacyIndiaDf.rename({'State Code':'state/ut'},axis=1)
literacyIndiaDf
```

```python
# Generating output csv file from datframe
literacyIndiaDf.to_csv('literacy-india.csv',index=False)
```
