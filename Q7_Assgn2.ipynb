```python
# Importing libraries
import pandas as pd
```

```python
# Reading language speakers data statewise for all states
JKlanguageDf = pd.read_excel('input/DDW-C17-0100.xlsx', names=['State Code','state/ut','L1_code','language','Persons_L1','Males_L1','Females_L1','L2_code','L2','Persons_L2','Males_L2','Females_L2','L3_code','L3','Persons_L3','Males_L3','Females_L3'])
JKlanguageDf = JKlanguageDf[JKlanguageDf['State Code']=='01']
JKlanguageDf
```

```python
# Preprocessing data and finding language wise speakers for mother tongue, 2nd language and 3rd language

JKlanguageDf1 = JKlanguageDf[['State Code','state/ut','language','Persons_L1']]
JKlanguageDf1 = JKlanguageDf1.dropna()


JKlanguageDf2 = JKlanguageDf[['L2','Persons_L2']]
JKlanguageDf2 = JKlanguageDf2.dropna()


JKlanguageDf3 = JKlanguageDf[['L3','Persons_L3']]
JKlanguageDf3 = JKlanguageDf3.dropna()


langDf = JKlanguageDf1[['language']]
langDf2 = JKlanguageDf2[['L2']]
langDf3 = JKlanguageDf3[['L3']]
langDf2 = langDf2.rename({'L2':'language'},axis=1)
langDf3 = langDf3.rename({'L3':'language'},axis=1)
langDf = pd.concat([langDf,langDf2,langDf3])
langDf = langDf.groupby('language').sum()
langDf = langDf.reset_index()
langDf['Persons_L1'] = 0
langDf['Persons_L2'] = 0
langDf['Persons_L3'] = 0


JKlanguageDf1 = pd.concat([JKlanguageDf1,langDf[['language','Persons_L1']]])
JKlanguageDf1 = JKlanguageDf1.groupby('language').sum()
JKlanguageDf1 = JKlanguageDf1.reset_index()


JKlanguageDf2 = JKlanguageDf2.rename({'L2':'language'},axis=1)
JKlanguageDf2 = pd.concat([JKlanguageDf2,langDf[['language','Persons_L2']]])
JKlanguageDf2 = JKlanguageDf2.groupby('language').sum()
JKlanguageDf2 = JKlanguageDf2.reset_index()


JKlanguageDf3 = JKlanguageDf3.rename({'L3':'language'},axis=1)
JKlanguageDf3 = pd.concat([JKlanguageDf3,langDf[['language','Persons_L3']]])
JKlanguageDf3 = JKlanguageDf3.groupby('language').sum()
JKlanguageDf3 = JKlanguageDf3.reset_index()


JKlanguageDf1 = pd.concat([JKlanguageDf1,JKlanguageDf2[['Persons_L2']]],axis=1)
JKlanguageDf1 = pd.concat([JKlanguageDf1,JKlanguageDf3[['Persons_L3']]],axis=1)
JKlanguageDf1['State Code'] = '01'
JKlanguageDf1['state/ut'] = 'JAMMU & KASHMIR'
JKlanguageDf = JKlanguageDf1[['State Code','state/ut','language','Persons_L1','Persons_L2','Persons_L3']]
JKlanguageDf
```
